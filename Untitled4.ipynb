```python
from sklearn.datasets import make_blobs, make_moons
import matplotlib.pyplot as plt

X, y = make_blobs(n_samples=100, centers=7 , cluster_std=1, n_features=2,
                  random_state=123)
print(X.shape)
print(X[0, :])
```

```python
plt.scatter(X[ :,0],X[:,1],c=y,alpha=0.7)
plt.title('dataset')
plt.show()
```

```python
x,z=make_moons(n_samples=200, noise=0.08, random_state=123)
print(x.shape)
```

```python
plt.scatter(x[ :,0],x[:,1],c=z,alpha=0.7)
plt.title('dataset')
plt.show()
plt.scatter
```

```python
from sklearn.cluster import AgglomerativeClustering
import numpy as np
import pandas as pd
agglo = AgglomerativeClustering(n_clusters=2)
agglo.fit(X)
mycolors={0:'Red',1:'green',2:'Blue'}
plt.scatter(X[ :,0],X[ :,1],c=pd.Series(agglo.labels_).apply(lambda x:mycolors[x]),alpha=0.7)
plt.title('dataset 1 agglomerative')
plt.show()
```

```python
agglo = AgglomerativeClustering(n_clusters=6)
agglo.fit(x)
mycolors={0:'Red',1:'green',2:'Blue',3:'yellow',4:'pink',5:'purple'}
plt.scatter(x[ :,0],x[ :,1],c=pd.Series(agglo.labels_).apply(lambda x:mycolors[x]),alpha=0.7)
plt.title('dataset 1 agglomerative')
```

```python
from scipy.cluster.hierarchy import linkage,dendrogram,fcluster
```

```python
mylinkage=linkage(x,method='median')
plt.figure(figsize=(40,20))
dendrogram(mylinkage)
plt.show()
```

```python
lables=fcluster(mylinkage,1,criterion='distance')
pd.Series(lables).value_counts()
```

### AgglomerativeClustering on iris dataset


```python
from sklearn import datasets
import pandas as pd
```

```python
iris=datasets.load_iris()
iris_dataset=iris.data
iris_dataset_pd=pd.DataFrame(iris.data,columns=iris.feature_names)
iris_dataset_pd.head()
```

```python
from sklearn.cluster import AgglomerativeClustering
import numpy as np
import pandas as pd
linkage=['complete','average','ward']
for idx,i in enumerate(linkage):
  plt.figure(idx)
  agglo = AgglomerativeClustering(n_clusters=3,affinity='euclidean',linkage=i)
  agglo.fit(iris_dataset_pd)
  mycolors={0:'Red',1:'green',2:'Blue'}
  plt.scatter(iris_dataset_pd.iloc[ :,2],iris_dataset_pd.iloc[ :,3],c=(agglo.labels_),alpha=0.7)
  plt.title('dataset 1 agglomerative'+''+i)
  plt.show()
```

```python
from scipy.cluster.hierarchy import linkage,dendrogram,fcluster
```

```python
mylinkage=linkage(iris_dataset_pd,method='ward')
plt.figure(figsize=(40,20))
dendrogram(mylinkage)
plt.show()
```

```python
lables=fcluster(mylinkage,6.7,criterion='distance')
pd.Series(lables).value_counts()
```
