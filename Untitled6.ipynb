```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import os
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs
%matplotlib inline
```

```python
x1, label1 = make_blobs(n_samples=100, n_features=2,centers=2,cluster_std=5,random_state=123)
plt.scatter(x1[:,0],x1[:,1],c=label1,alpha=0.7)
plt.title("Dataset #1 : Original")
plt.show()
```

```python
x2, label2 = make_blobs(n_samples=100,n_features=2,centers=3,cluster_std=1,random_state=321)
plt.scatter(x2[:,0],x2[:,1],c=label2,alpha=0.7)
plt.title("Dataset # : Original")
```

```python
kmeans = KMeans(n_clusters=2,random_state=123)
kmeans.fit(x1)
mycolours = {0:"red",1:"green",2:"blue"}
plt.scatter(x1[:,0],x1[:,1],c=pd.Series(kmeans.labels_).apply(lambda x: mycolours[x]),alpha=0.7)
plt.title("Dataset #1 : K Means")
```

```python
kmeans = KMeans(n_clusters=3,random_state=123)
kmeans.fit(x1)
mycolours = {0:"red",1:"green",2:"blue"}
plt.scatter(x1[:,0],x1[:,1],c=pd.Series(kmeans.labels_).apply(lambda x: mycolours[x]),alpha=0.7)
plt.title("Dataset #1 : K Means")
```

```python
kmeans = KMeans(n_clusters=2,random_state=123)
kmeans.fit(x2)
mycolours = {0:"red",1:"green",2:"blue"}
plt.scatter(x2[:,0],x2[:,1],c=pd.Series(kmeans.labels_).apply(lambda x: mycolours[x]),alpha=0.7)
plt.title("Dataset #2 : K Means")
```

```python
kmeans = KMeans(n_clusters=3,random_state=123)
kmeans.fit(x2)
mycolours = {0:"red",1:"green",2:"blue"}
plt.scatter(x2[:,0],x2[:,1],c=pd.Series(kmeans.labels_).apply(lambda x: mycolours[x]),alpha=0.7)
plt.title("Dataset #2 : K Means")
```

```python
print(kmeans.inertia_)
```

```python
a=[i for i in range(2,11)]
b=[]
for i in a:
  kmeans = KMeans(n_clusters=i,random_state=123)
  kmeans.fit(x2)
  b.append(kmeans.inertia_)
plt.plot(a,b,marker="+")  
```
